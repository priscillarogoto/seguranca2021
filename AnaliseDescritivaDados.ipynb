# Usando atributos "Quadris" e "Cintura" para gerar nova coluna no dataset com dados de RCQ (Relação Cintura/Quadris)


## RCQ.png

```python
df1['RCQ'] = df1['Cintura'] / df1['Quadris']
```

```python
print (df1)
```

# Criando função que atribui a classificação de gordura corporal de acordo com a tabela de RCQ:


```python
def classificar_gordura_corporal(RCQ, Genero):
    if Genero == 1: # masculino
        if RCQ < 0.95:
            return 'Baixo'
        elif RCQ >= 0.96 and RCQ < 1.00:
            return 'Moderado'
        else:
            return 'Alto'
    else: # feminino
        if RCQ < 0.80:
            return 'Baixo'
        elif RCQ >= 0.81 and RCQ < 0.85:
            return 'Moderado'
        else:
            return 'Alto'
```

# Aplicando a função ao DataFrame para classificar a gordura corporal de cada indivíduo:


```python
df1['Classificacao'] = df1.apply(lambda row: classificar_gordura_corporal(row['RCQ'], row['Genero']), axis=1)
```

```python
print(df1)
```

# Agrupando os dados pelo gênero e pela classificação de gordura corporal:


```python
dados_agrupados = df1.groupby(['Genero', 'Classificacao'])['RCQ'].count()
```

# Criando um gráfico de barras empilhadas com os dados agrupados:


```python
dados_agrupados.unstack().plot(kind='bar', stacked=True)
plt.xlabel('Gênero')
plt.ylabel('Número de indivíduos')
plt.title('Classificação de gordura corporal')
plt.legend(title='Classificação', loc='upper right')
plt.show()
```
